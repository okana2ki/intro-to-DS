## 関数の定義と使用例

Pythonで関数を定義するには`def`キーワードを使います。関数は特定のタスクを実行するコードのブロックをまとめたもので、何度も同じ処理を記述する手間を省き、コードを整理し、再利用しやすくするために使われます。

以下の例では、2つの数値を足し算するシンプルな関数`add_numbers`を定義し、それを使用する方法を示します。

````python
def add_numbers(a, b):
    """2つの数値を足し算して結果を返します。"""
    sum_result = a + b
    return sum_result

# 関数の使用例
number1 = 10
number2 = 5

# add_numbers関数を呼び出して、結果を格納
result = add_numbers(number1, number2)

# 結果を表示
print(f"{number1} と {number2} を足し算すると {result} になります。")

# 別の値で関数を呼び出すこともできます
print(f"3 と 7 を足し算すると {add_numbers(3, 7)} になります。")
````

### コードの説明:

*   `def add_numbers(a, b):`:
    *   `def` は「ここから関数を定義しますよ」という合図です。
    *   `add_numbers` は関数の名前です。何をする関数なのかがわかるような名前にするのが一般的です。
    *   `(a, b)` は「引数（ひきすう）」と呼ばれるもので、関数に渡す値のことです。この関数では、`a` と `b` の2つの数値を受け取ります。
    *   行の最後には必ずコロン`:`をつけます。
*   `"""2つの数値を足し算して結果を返します。"""`:
    *   これは「docstring（ドックストリング）」と呼ばれるもので、関数の説明を書くために使われます。コードを読む人が関数が何をするのかを理解するのに役立ちます。
*   `sum_result = a + b`:
    *   関数内で実行される処理です。ここでは、受け取った `a` と `b` を足し算して、その結果を `sum_result` という変数に代入しています。
*   `return sum_result`:
    *   `return` は「関数の結果としてこの値を返しますよ」という指示です。この関数は計算結果の `sum_result` を返します。
*   `result = add_numbers(number1, number2)`:
    *   これが関数の「呼び出し」です。関数名と、定義した引数に対応する値をカッコの中に入れて記述します。この呼び出しによって、関数内の処理が実行され、`return`で返された値が`result`変数に代入されます。
